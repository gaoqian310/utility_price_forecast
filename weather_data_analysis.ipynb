```python
# Import pandas library
import os
import pandas as pd
```

```python
# Read the CSV file into a DataFrame object
df = pd.read_csv("weather_data/Calgary_2020.csv",encoding = "cp1252")
```

```python
# View the first 5 rows of the DataFrame
df.head()
```

```python
# View the summary statistics of the numerical columns
df.describe()
```

```python
# View the data types and missing values of each column
df.info()
```

```python
# View the distribution of a single column using a histogram
df["Temp (°C)"].hist()
```

```python
# View the correlation between two columns using a scatter plot
df.plot.scatter(x="Temp (°C)", y="Dew Point Temp (°C)")

# View the relationship between multiple columns using a pair plot
import seaborn as sns # another useful library for data visualization
sns.pairplot(df[["Temp (°C)", "Dew Point Temp (°C)", "Rel Hum (%)", "Wind Spd (km/h)"]])
```
