```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('merged_expanded_data.csv', parse_dates=['DateTime'])
```

```python
data.describe()
```

```python
# Convert columns that should be numeric
numeric_columns = ['Temp (°C)', 'Wind Chill', 'Rel Hum (%)', 'Stn Press (kPa)', 'NG_price', 'pool_price']
for column in numeric_columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')

# Drop rows with NaN values after conversion
data.dropna(subset=numeric_columns, inplace=True)

# Summary statistics
print("Summary Statistics:")
print(data.describe())

# Correlation matrix
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

# Distribution plots
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(data['Temp (°C)'], bins=20, kde=True, color='skyblue')
plt.title("Temperature Distribution")

plt.subplot(2, 2, 2)
sns.histplot(data['Wind Chill'], bins=20, kde=True, color='salmon')
plt.title("Wind Chill Distribution")

plt.subplot(2, 2, 3)
sns.histplot(data['Rel Hum (%)'], bins=20, kde=True, color='green')
plt.title("Humidity Distribution")

plt.subplot(2, 2, 4)
sns.histplot(data['Stn Press (kPa)'], bins=20, kde=True, color='purple')
plt.title("Pressure Distribution")

plt.tight_layout()
plt.show()

# Scatter plot: Natural Gas Price vs. Electricity Price
plt.figure(figsize=(8, 6))
sns.scatterplot(x='NG_price', y='pool_rice', data=data, color='orange')
plt.title("Natural Gas Price vs. Electricity Price")
plt.xlabel("Natural Gas Price")
plt.ylabel("Electricity Price")
plt.show()
```
