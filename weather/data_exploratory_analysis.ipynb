```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Specify the folder path where your CSV files are located
folder_path = 'aggregated_weather_data'

# Get a list of all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.lower().endswith('_aggregated_weather.csv')]
```

```python
# Read each CSV file and create separate plots
for csv_file in csv_files:
    city_name = os.path.splitext(csv_file)[0].split('_')[0]  # Extract city name
    df = pd.read_csv(os.path.join(folder_path, csv_file))
    
    # Summary statistics
    print(f"Summary statistics for {city_name}:")
    print(df.describe())
    
    # Heatmap to visualize missing values
    plt.figure(figsize=(8, 6))
    sns.heatmap(df.isnull(), cmap='coolwarm', cbar=False)
    plt.title(f"Missing Values Heatmap for {city_name}")
    plt.show()

    # Time series plot for temperature
    plt.figure(figsize=(10, 6))
    df['Date/Time (LST)'] = pd.to_datetime(df['Date/Time (LST)'])
    df.set_index('Date/Time (LST)', inplace=True)
    df['Temp (°C)'].plot()
    plt.xlabel("Date")
    plt.ylabel("Temperature (°C)")
    plt.title(f"Temperature Variation in {city_name}")
    plt.grid(True)
    plt.show()

    # Dew Point Temperature Variation
    plt.figure(figsize=(10, 6))
    df['Dew Point Temp (°C)'].plot()
    plt.xlabel("Date")
    plt.ylabel("Dew Point Temperature (°C)")
    plt.title(f"Dew Point Temperature Variation in {city_name}")
    plt.grid(True)
    plt.show()

    # Relative Humidity Trends
    plt.figure(figsize=(10, 6))
    df.groupby('Month')['Rel Hum (%)'].mean().plot(kind='bar')
    plt.xlabel("Month")
    plt.ylabel("Relative Humidity (%)")
    plt.title(f"Relative Humidity Trends in {city_name}")
    plt.grid(True)
    plt.show()

    # Pressure Trends
    plt.figure(figsize=(10, 6))
    df['Stn Press (kPa)'].plot()
    plt.xlabel("Date")
    plt.ylabel("Stn Press (kPa)")
    plt.title(f"Station Pressure Variation in {city_name}")
    plt.grid(True)
    plt.show()

     # Wind Chill Trends
    plt.figure(figsize=(10, 6))
    df.groupby('Month')['Wind Chill'].mean().plot(kind='bar')
    plt.xlabel("Month")
    plt.ylabel("Wind Chill")
    plt.title(f"Wind Chill Trends in {city_name}")
    plt.grid(True)
    plt.show()

    print("\n")  # Add a separator between cities
```

```python
# Initialize a dictionary to store common weather conditions
common_weather = {}

# Read each CSV file and analyze weather conditions
for csv_file in csv_files:
    city_name = os.path.splitext(csv_file)[0].split('_')[0]  # Extract city name
    df = pd.read_csv(os.path.join(folder_path, csv_file))

    # Combine "clear" and "mainly clear" into a single category
    df['Weather'] = df['Weather'].replace({'Mainly Clear': 'Clear'})

    # Combine "cloudy" and "mainly cloudy" into a single category
    df['Weather'] = df['Weather'].replace({'Mostly Cloudy': 'Cloudy'})
    
    # Count occurrences of each weather condition
    weather_counts = df['Weather'].value_counts()

    # Get the top 10 most common weather conditions
    top_10_weather = weather_counts.nlargest(10)
    
    # Store the result in the dictionary
    common_weather[city_name] = top_10_weather

    # Plot the most common weather
    plt.figure(figsize=(8, 5))
    top_10_weather.plot(kind='bar', color='skyblue')
    plt.xlabel("Weather Condition")
    plt.ylabel("Frequency")
    plt.title(f"Most Common Weather in {city_name}")
    plt.grid(axis='y')
    plt.show()

# Print the summary
for city, weather in common_weather.items():
    print(f"Most common weather in {city}: {weather}")
```
