```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a DataFrame object
df = pd.read_csv("https://dd.weather.gc.ca/climate/observations/hourly/csv/AB/climate_hourly_AB_3031092_2020_P1H.csv",encoding = "cp1252")
# df = pd.read_csv("weather_data/Calgary_2020.csv",encoding = "cp1252")
```

```python
# View the first 5 rows of the DataFrame
df.head()
```

```python
# View the summary statistics of the numerical columns
df.describe()
```

```python
# View the data types and missing values of each column
df.info()
```

```python
# Remove empty columns
df = df.drop(columns=['Temp Flag','Dew Point Temp Flag','Rel Hum Flag','Precip. Amount (mm)','Precip. Amount Flag','Wind Dir Flag',
                      'Wind Spd Flag','Visibility Flag','Stn Press Flag','Hmdx','Hmdx Flag','Wind Chill Flag'])

df.info()
```

```python
sns.histplot(df["Temp (°C)"], kde=True)
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.title("Distribution of Temperature in 2020")
plt.show()
```

```python
df["Date"] = pd.to_datetime(df[["Year", "Month", "Day"]])

# Group by date and get the maximum temperature
df_max = df.groupby("Date")["Temp (°C)"].max()

# Group by date and get the minimum temperature
df_min = df.groupby("Date")["Temp (°C)"].min()

# Plot the data
plt.plot(df_max, color='r', label='Maximum Temperature')
plt.plot(df_min, color='b', label='Minimum Temperature')
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Maximum and Minimum Temperature of the Day in 2020")
plt.grid()
plt.legend()
plt.show()
```

```python
# Calculate the weather counts
weather_counts = df["Weather"].value_counts()

# Get the top x weather conditions
top_weather_counts = weather_counts.head(13)

# Plot the top weather summary
plt.figure(figsize=(10, 6))
sns.barplot(x=top_weather_counts.index, y=top_weather_counts.values, palette="coolwarm")
plt.xlabel("Weather")
plt.ylabel("Hours")
plt.title("Top 13 Weather Conditions in 2020")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y")
plt.show()
```

```python
# Create a pivot table to aggregate temperature data by month
monthly_weather = df.pivot_table(index=df["Date"].dt.month, columns=df["Date"].dt.year, values="Temp (°C)", aggfunc="mean")

# Plot the heat map
plt.figure(figsize=(10, 6))
sns.heatmap(monthly_weather, cmap="coolwarm", annot=True, fmt=".1f", cbar_kws={"label": "Temperature (°C)"})
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Monthly Weather Heat Map in 2020")
plt.show()
```

```python
# Create a pivot table to aggregate temperature data by month
monthly_weather = df.pivot_table(index=df["Date"].dt.month, columns=df["Date"].dt.day, values="Temp (°C)", aggfunc="mean")

# Plot the heat map without displaying values
plt.figure(figsize=(10, 6))
sns.heatmap(monthly_weather, cmap="coolwarm", annot=False, cbar_kws={"label": "Temperature (°C)"})
plt.xlabel("Day of the Month")
plt.ylabel("Month")
plt.title("Daily Weather Heat Map (2020)")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame object
df = pd.read_csv("https://dd.weather.gc.ca/climate/observations/hourly/csv/AB/climate_hourly_AB_3031092_2020_P1H.csv", encoding="cp1252")

# Convert the date columns to datetime format
df["Date"] = pd.to_datetime(df[["Year", "Month", "Day"]])

# Convert the "Time (LST)" column to datetime objects
df["Time"] = pd.to_datetime(df["Time (LST)"], format="%H:%M").dt.time

# Filter data for peak hours (16:00-21:00) on the same date
peak_hour_df = df[(df["Time"] >= pd.to_datetime("16:00").time()) & (df["Time"] < pd.to_datetime("21:00").time())]

# Calculate the average temperature during peak hours
peak_hour_average_temp = peak_hour_df.groupby("Date")["Temp (°C)"].mean()

# Filter data for off-peak hours (21:00-16:00 next day, including the date+1)
offpeak_hour_df = df[((df["Time"] >= pd.to_datetime("21:00").time()) | (df["Time"] < pd.to_datetime("16:00").time()))]

# Calculate the average temperature during off-peak hours
offpeak_hour_average_temp = offpeak_hour_df.groupby("Date")["Temp (°C)"].mean()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(peak_hour_average_temp, label="Peak Hours (16:00-21:00)", color="r")
plt.plot(offpeak_hour_average_temp, label="Off-Peak Hours (21:00-16:00 next day)", color="b")
plt.xlabel("Date")
plt.ylabel("Average Temperature (°C)")
plt.title("Average Temperature during Peak and Off-Peak Hours")
plt.grid()
plt.legend()
plt.show()
```

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame object
df = pd.read_csv("https://dd.weather.gc.ca/climate/observations/hourly/csv/AB/climate_hourly_AB_3031092_2020_P1H.csv", encoding="cp1252")

# Convert the date columns to datetime format
df["Date"] = pd.to_datetime(df[["Year", "Month", "Day"]])

# Convert the "Time (LST)" column to datetime objects
df["Time"] = pd.to_datetime(df["Time (LST)"], format="%H:%M").dt.time

# Filter data for January
january_df = df[df["Date"].dt.month == 1]

# Calculate the average temperature during peak hours in January
january_peak_hour_df = january_df[(january_df["Time"] >= pd.to_datetime("16:00").time()) & (january_df["Time"] < pd.to_datetime("21:00").time())]
january_peak_hour_average_temp = january_peak_hour_df.groupby("Date")["Temp (°C)"].mean()

# Calculate the average temperature during off-peak hours in January
january_offpeak_hour_df = january_df[((january_df["Time"] >= pd.to_datetime("21:00").time()) | (january_df["Time"] < pd.to_datetime("16:00").time()))]
january_offpeak_hour_average_temp = january_offpeak_hour_df.groupby("Date")["Temp (°C)"].mean()

# Plot the data for January
plt.figure(figsize=(10, 6))
plt.plot(january_peak_hour_average_temp, label="Peak Hours (16:00-21:00)", color="r")
plt.plot(january_offpeak_hour_average_temp, label="Off-Peak Hours (21:00-16:00 next day)", color="b")
plt.xlabel("Date and Time")
plt.ylabel("Average Temperature (°C)")
plt.title("Average Temperature during Peak and Off-Peak Hours in January")
plt.grid()
plt.legend()
plt.show()

```
