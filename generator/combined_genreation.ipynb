```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from each file
coal_df = pd.read_csv('Daily_Coal_Generation.csv', parse_dates=['Date'])
gas_df = pd.read_csv('daily_gas_generation.csv', parse_dates=['Date'])
solar_df = pd.read_csv('Daily_Solar_Generation.csv', parse_dates=['Date'])
wind_df = pd.read_csv('Daily_wind_generation.csv', parse_dates=['Date'])

# Rename the 'Generation' column to match the generator type
coal_df.rename(columns={'Clean_Coal_Generation': 'Coal'}, inplace=True)
gas_df.rename(columns={'Daily Gas Generation': 'Gas'}, inplace=True)
solar_df.rename(columns={'Clean_Solar_Generation': 'Solar'}, inplace=True)
wind_df.rename(columns={'Daily Wind Generation': 'Wind'}, inplace=True)

# Merge the dataframes on the 'Date' column
combined_df = coal_df.merge(gas_df, on='Date').merge(solar_df, on='Date').merge(wind_df, on='Date')

# Save the combined data to a new file
combined_df.to_csv('combined_daily_generation.csv', index=False)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(combined_df['Date'], combined_df['Coal'], label='Coal')
plt.plot(combined_df['Date'], combined_df['Gas'], label='Gas')
plt.plot(combined_df['Date'], combined_df['Solar'], label='Solar')
plt.plot(combined_df['Date'], combined_df['Wind'], label='Wind')
plt.xlabel('Date')
plt.ylabel('Generation (in MW)')
plt.title('Daily Generation Comparison')
plt.legend()
plt.show()
```
